```python
def products_recommend(num_of_acne:int, skin_type: str) -> dict:
    if num_of_acne <= 5:
        if skin_type == "Normal":
            product_names = df_normal["name"].iloc[:90].sample(4, random_state=15)
        elif skin_type == "Dry":
            product_names = df_dry["name"].iloc[:90].sample(4, random_state=15)
        elif skin_type == "Oily":
            product_names = df_oily["name"].iloc[:90].sample(4, random_state=15)
    elif num_of_acne > 5 and num_of_acne <= 10:
        if skin_type == "Normal":
            product_names = df_normal["name"].iloc[90:200].sample(4, random_state=15)
        elif skin_type == "Dry":
            product_names = df_dry["name"].iloc[90:200].sample(4, random_state=15)
        elif skin_type == "Oily":
            product_names = df_oily["name"].iloc[90:200].sample(4, random_state=15)
    else:
        if skin_type == "Normal":
            product_names = df_normal["name"].iloc[200:].sample(4, random_state=15)
        elif skin_type == "Dry":
            product_names = df_dry["name"].iloc[200:].sample(4, random_state=15)
        elif skin_type == "Oily":
            product_names = df_oily["name"].iloc[200:].sample(4, random_state=15)
    
    if skin_type == "Normal":
        return {
            "product_name" : product_names.to_list(),
            "product_image" : df_normal[df_normal["name"].isin(product_names.to_list())]["img_source"].to_list()
        }
    elif skin_type == "Dry":
        return {
            "product_name" : product_names.to_list(),
            "product_image" : df_dry[df_dry["name"].isin(product_names.to_list())]["img_source"].to_list()
        }
    else:
        return {
            "product_name" : product_names.to_list(),
            "product_image" : df_oily[df_oily["name"].isin(product_names.to_list())]["img_source"].to_list()
        }
    
```
